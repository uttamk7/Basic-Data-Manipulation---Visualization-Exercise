### 53. Convert the below Series to pandas datetime : 

DoB = pd.Series(["07Sep59","01Jan55","15Dec47","11Jul42"])

Make sure that the year is 19XX not 20XX

```python
import pandas as pd
DoB = pd.Series(["07Sep59","01Jan55","15Dec47","11Jul42"])

convert_date = lambda date: pd.to_datetime(date, format='%d%b%y') - pd.DateOffset(years=100) if pd.to_datetime(date, format='%d%b%y').year >= 2000 else pd.to_datetime(date, format='%d%b%y')

DoB_datetime = DoB.apply(convert_date)

print(DoB_datetime)
```

### 54. Write a Python program to get days between two dates. 

```python
from datetime import datetime

def days_between_dates(date1_str, date2_str, date_format="%Y-%m-%d"):
    date1 = datetime.strptime(date1_str, date_format)
    date2 = datetime.strptime(date2_str, date_format)
    
    delta = date2 - date1
    
    return delta.days


date1_str = "2023-05-01"
date2_str = "2024-05-22"

days_between = days_between_dates(date1_str, date2_str)
print(f"Days between {date1_str} and {date2_str} are: {days_between}days")
```

### 80. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']

top_countries = df.groupby('Country')['Revenue'].sum().nlargest(3).index

top_countries_df = df[df['Country'].isin(top_countries)]

fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

for ax, country in zip(axes, top_countries):
    country_df = top_countries_df[top_countries_df['Country'] == country]
    ax.scatter(country_df['Quantity'], country_df['UnitPrice'], alpha=0.7)
    ax.set_title(f'Scatter Plot for {country}')
    ax.set_xlabel('Quantity')
    ax.set_ylabel('Unit Price')
    ax.grid()

plt.tight_layout()
plt.show()
```
